## Observations and Insights 
1. From all the treatments, Propriva run the least experiments
2. The Capomulin Regimen shows a tumor volume reduction over time
3. There is a high correlation between the average tumor volume and the mice weight on the Capomulin Regimen
4. The regimens with the greatest number of measurements, Capomulin and Ramican, also have the lower variation and standard deviation

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
Study_dataset = pd.merge(mouse_metadata,study_results, how= "left", on="Mouse ID")

# Display the data table for preview
Study_dataset.head()
```

```python
# Checking the number of mice.
Study_dataset["Mouse ID"].value_counts()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = Study_dataset.loc[Study_dataset.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
Study_dataset.loc[Study_dataset["Mouse ID"] == "g989",:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_dataset = Study_dataset[Study_dataset["Mouse ID"].isin(duplicate_mice) == False]
clean_dataset
```

```python
# Checking the number of mice in the clean DataFrame.
clean_dataset["Mouse ID"].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
Regimen_df = clean_dataset.groupby("Drug Regimen")

# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean = Regimen_df["Tumor Volume (mm3)"].mean()
median = Regimen_df["Tumor Volume (mm3)"].median()
variance = Regimen_df["Tumor Volume (mm3)"].var()
Standard_deviation = Regimen_df["Tumor Volume (mm3)"].std()
SEM = Regimen_df["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
Summary_statistics = pd.DataFrame({"mean":mean, "median":median, "variance":variance, "Standard Deviation":Standard_deviation, "SEM":SEM})
Summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = pd.DataFrame(clean_dataset.groupby("Mouse ID")["Timepoint"].max().sort_values()).reset_index().rename(columns={"Timepoint": "Last Timepoint"})

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_volume = pd.merge(clean_dataset, last_timepoint, on = "Mouse ID")
tumor_volume
```

```python
# Put treatments into a list for for loop (and later for plot labels)
Treatments = ("Ramicane", "Capomulin", "Infubinol", "Ceftamin")

# Create empty list to fill with tumor vol data (for plotting)
Tumor_volume=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in Treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    regimen_df = tumor_volume.loc[tumor_volume["Drug Regimen"] == regimen]
    
    # add subset
    last_volume = regimen_df.loc[regimen_df["Timepoint"]==regimen_df["Last Timepoint"]]
    last_volumes = last_volume["Tumor Volume (mm3)"]
    Tumor_volume.append(last_volumes)
    
    # Determine outliers using upper and lower bounds
    volume = last_volume["Tumor Volume (mm3)"]
    quartiles = volume.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    IQR = upperq - lowerq
    lowerb = lowerq - (1.5*IQR)
    upperb = upperq + (1.5*IQR)
    Outliers = (volume.loc[(last_volume["Tumor Volume (mm3)"] >= upperb) |
                (last_volume['Tumor Volume (mm3)'] <= lowerb)]).count()
    print(f"{regimen}: {Outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
#highlight any potential outliers in the plot by changing their color and style
outliers_props = dict(markerfacecolor="yellow", markersize=15)
fig1, ax1 = plt.subplots(figsize=(18, 6))
plt.boxplot(Tumor_volume, flierprops=outliers_props)
plt.xticks([1,2,3,4],["Ramicane", "Capomlin", "Infubinol", "Ceftamin"], fontsize = 14)
plt.yticks(fontsize =14)
plt.xlabel("Drug Regimens", fontsize =16)
plt.ylabel("Last Tumor Volume", fontsize =16)
plt.title("Last Tumor Volume by Drug Regimen", fontsize = 20)

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#Select a random mouse from the Capomulin Treatment
Capomulin_mice = clean_dataset.loc[clean_dataset["Drug Regimen"] == "Capomulin"]
Capomulin_mice.sample()
Capomulin_mouse = clean_dataset.loc[clean_dataset["Mouse ID"] == "i738"]

#Line Chart
fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.plot(Capomulin_mouse["Timepoint"], Capomulin_mouse["Tumor Volume (mm3)"], marker = "o", markersize=9)
plt.xlabel("Time Point", fontsize = 16)
plt.ylabel("Tumor Volume (mm3)", fontsize = 16)
plt.xticks(size = 14)
plt.yticks(size = 14)
plt.title("Tumor Volume vs Time for a mouse treated with Capomulin", fontsize = 20)
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
avg_vol = pd.DataFrame (Capomulin_mice.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()).rename(columns={"Tumor Volume (mm3)": "Average Volume"}).reset_index()
Capomulin_avg_vol = pd.merge(Capomulin_mice, avg_vol, on = "Mouse ID")
plt.scatter(Capomulin_avg_vol["Weight (g)"], Capomulin_avg_vol["Average Volume"])
plt.xlabel("Weight (g)")
plt.ylabel("Average tumor volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight for the Capomulin regimen")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and average tumor volume for the Capomulin regimen 
corr_coef = np.corrcoef(Capomulin_avg_vol["Weight (g)"], Capomulin_avg_vol["Average Volume"])
print(corr_coef)
#Calculate linear regression model for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept, rvalue, pvalue, stderr) = linregress(Capomulin_avg_vol["Weight (g)"], Capomulin_avg_vol["Average Volume"])
Lin_reg = (Capomulin_avg_vol["Weight (g)"]) * slope + intercept
#Plot the linear regression model on top of the previous scatter plot.
plt.figure(figsize=(25, 15))
plt.scatter(Capomulin_avg_vol["Weight (g)"], Capomulin_avg_vol["Average Volume"], s = 150)
plt.plot((Capomulin_avg_vol["Weight (g)"]), Lin_reg, "r-")
plt.xlabel("Weight (g)", fontsize = 22)
plt.ylabel("Average tumor volume (mm3)", fontsize = 22)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.title("Average Tumor Volume vs Mouse Weight for the Capomulin Regimen", fontsize = 26)

#Show regression equation
reg_eq= "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(reg_eq, xy=(22,37), color="black", fontsize = 32)
coef= "Correlation Coefficient = 0.83"
plt.annotate(coef, xy=(21,36), color = "black", fontsize =32)
plt.show()
```
